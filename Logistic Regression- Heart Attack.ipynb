```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
data = df.copy()
X = data.iloc[:,0:13]  #independent columns
y = data.iloc[:,-1]    #target column i.e price range


bestfeatures = SelectKBest(score_func=chi2, k=10)
(X < 0).sum()  



```

```python
X_positive = X.abs()  # Use absolute values to ensure all are non-negative
fit = bestfeatures.fit(X_positive, y)

dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 

featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(12,'Score'))  #print 10 best features
```

```python
from sklearn.ensemble import ExtraTreesClassifier

model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(13).plot(kind='barh')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
log_reg = LogisticRegression()

log_reg = LogisticRegression(max_iter=1100)
log_reg.fit(X_train, y_train)


y_pred = log_reg.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Output the results
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

1. Accuracy: 0.84
Interpretation: The accuracy of 84.45% means that the model correctly classified 84.45% of all instances in the test set. This is a good overall indicator, but accuracy alone doesn't give the full picture, especially if the classes are imbalanced.


2. Confusion Matrix:
[[ 95  16]
 [ 21 106]]
True Positives (TP): 106  Correct predictions where the target was 1.

True Negatives (TN): 95  Correct predictions where the target was 0.

False Positives (FP): 16  Incorrect predictions where the target was predicted as 1, but was actually 0.

False Negatives (FN): 21  Incorrect predictions where the target was predicted as 0, but was actually 1.

Interpretation: The confusion matrix shows that the model is performing well, but there are some false positives and false negatives. The model is slightly more conservative, with fewer false positives (16), and more false negatives (21), which means it sometimes misses cases that are positive (i.e., heart attacks).



3. Precision, Recall, and F1-Score:

For Class 0 (Non-Heart Attack):

Precision: 0.82  Out of all predicted non-heart attack cases, 82% were actually non-heart attack cases.
Recall: 0.86  Out of all actual non-heart attack cases, the model identified 86% correctly.
F1-Score: 0.84  A balance between precision and recall.

For Class 1 (Heart Attack):
Precision: 0.87  Out of all predicted heart attack cases, 87% were correct predictions.
Recall: 0.83  Out of all actual heart attack cases, the model identified 83% correctly.
F1-Score: 0.85  A good balance between precision and recall for heart attack cases.


4. Macro and Weighted Averages:

Macro Avg: The simple average of precision, recall, and F1-score across both classes. It's useful when you want to treat all classes equally, regardless of their support.

Precision: 0.84, Recall: 0.85, F1-Score: 0.84  These values show a good balance between classes.

Weighted Avg: The weighted average takes into account the support (the number of true instances for each class) when computing the average. Since the support for both classes is fairly balanced, the weighted averages closely resemble the macro averages.


5. Key Insights:
The model performs similarly well on both classes, though it is slightly better at predicting non-heart attack cases (Class 0) in terms of recall and slightly better at predicting heart attack cases (Class 1) in terms of precision.

The F1-scores of 0.84 for non-heart attack cases and 0.85 for heart attack cases show that the model strikes a good balance between precision and recall for both classes, which is essential in a medical diagnosis setting where false negatives and false positives can have significant implications.
